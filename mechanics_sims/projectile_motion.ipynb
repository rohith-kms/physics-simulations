```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Projectile Motion variables

g = 10 #m/s^2
missile_vs = [75,100,150] #m/s
theta = np.pi/6 # 30 degrees
```

```python
def trajectory(x, missile_type, theta):
    
    v = missile_vs[missile_type]
    
    return x*np.tan(theta) - 0.5*g*x**2/(v**2*(np.cos(theta))**2)
```

```python
xs = np.linspace(0,500,1000)

plt.plot(trajectory(xs,1,theta))
```
