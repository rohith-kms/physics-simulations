### The following is a simulation to plot the x coordinates of the two masses as a function of time.

```python
m1 = 10 # Kg
m2 = 5 # Kg
k = 0.001 # N/m
F = 1 # N
dt = 0.001 # s
```

```python
# Spring force

def F_s(x1,x2):
    
    return k*(x2-x1-1)
```

```python
# Initialise x values and next step

# Natural length of the spring is 1 units

x1_init = 0
x2_init = 1

x1_next = x1_init
x2_next = x2_init + 0.5*F/m2*(dt)*(dt)

x2_next
```

```python
# main loop parameters

timesteps = 600000

x1s = np.zeros(timesteps)
x2s = np.zeros(timesteps)

x1s[0] = x1_init
x1s[1] = x1_next
x2s[0] = x2_init
x2s[1] = x2_next    
```

```python
# main loop

for t in range(2,timesteps):
    
    x1s[t] = 2*x1s[t-1] - x1s[t-2] + F_s(x1s[t-1],x2s[t-1])/m1*dt*dt
    x2s[t] = 2*x2s[t-1] - x2s[t-2] + (F - F_s(x1s[t-1],x2s[t-1]))/m2*dt*dt
```

```python
# Position of COM

com_xs = (m1*x1s+m2*x2s)/(m1+m2)
```

```python
# Positions of the masses

fig,ax = plt.subplots(figsize=(14,12))

times = np.linspace(0,timesteps*dt,timesteps)

ax.plot(times, x1s)
ax.plot(times, x2s)
ax.plot(times, com_xs, '--')
```

```python
# About COM

fig,ax = plt.subplots(figsize=(14,12))

ax.plot(x1s-com_xs)
ax.plot(x2s-com_xs)
ax.plot(com_xs-com_xs, '--')
```
